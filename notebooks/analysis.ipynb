```python
# Sample data for testing
sample_books_data = {
    '1': {'title': 'Book A', 'author': 'Author X', 'genre': 'Fiction'},
    '2': {'title': 'Book B', 'author': 'Author Y', 'genre': 'Mystery'},
    '3': {'title': 'Book C', 'author': 'Author Z', 'genre': 'Science Fiction'},
}

sample_loans_data = [
    {'book_number': '1', 'date_of_loan': 44930, 'date_of_return': 44940},
    {'book_number': '2', 'date_of_loan': 44935, 'date_of_return': 44945},
    {'book_number': '3', 'date_of_loan': 44940, 'date_of_return': 44950},
    {'book_number': '1', 'date_of_loan': 44950, 'date_of_return': 44965},
]

valid_loans = sample_loans_data

# Function to calculate the popularity of books based on the number of days borrowed
def calculate_popularity(valid_loans):
    book_popularity = {}
    for loan in valid_loans:
        book_number = loan['book_number']
        days_borrowed = loan['date_of_return'] - loan['date_of_loan']
        if book_number not in book_popularity:
            book_popularity[book_number] = days_borrowed
        else:
            book_popularity[book_number] += days_borrowed
    return book_popularity

# Function to generate a popularity report of book genres
def generate_genre_popularity_report(books_data, valid_loans):
    genre_popularity = {}
    genre_book_count = {}

    # Calculate popularity and count for each genre
    for loan in valid_loans:
        book_number = loan['book_number']
        genre = books_data.get(book_number, {}).get('genre', 'Unknown')
        days_borrowed = loan['date_of_return'] - loan['date_of_loan']

        # Update genre popularity
        if genre not in genre_popularity:
            genre_popularity[genre] = days_borrowed
        else:
            genre_popularity[genre] += days_borrowed

        # Count books in each genre
        if genre not in genre_book_count:
            genre_book_count[genre] = 1
        else:
            genre_book_count[genre] += 1

    return genre_popularity, genre_book_count

# Function to calculate loan statistics
def calculate_loan_statistics(valid_loans):
    total_loan_days = 0
    total_loan_count = len(valid_loans)
    total_late_loan_days = 0
    total_late_loan_count = 0

    # Calculate total loan days and total late loan days
    for loan in valid_loans:
        days_borrowed = loan['date_of_return'] - loan['date_of_loan']
        total_loan_days += days_borrowed
        if days_borrowed > 14:  # Books returned after 14 days are considered late
            total_late_loan_days += (days_borrowed - 14)
            total_late_loan_count += 1

    # Calculate average loan days and average late loan days
    average_loan_days = total_loan_days / total_loan_count if total_loan_count > 0 else 0
    average_late_loan_days = total_late_loan_days / total_late_loan_count if total_late_loan_count > 0 else 0

    # Calculate the percentage proportion of books returned late
    percentage_returned_late = (total_late_loan_count / total_loan_count) * 100 if total_loan_count > 0 else 0

    return total_loan_days, total_loan_count, average_loan_days, \
        total_late_loan_days, total_late_loan_count, average_late_loan_days, percentage_returned_late

# Main code for testing

# Task 1: Test the book popularity analysis
print("\nTask 1: Book Popularity Analysis\n")
book_popularity = calculate_popularity(sample_loans_data)
sorted_books = sorted(sample_books_data.items(), key=lambda x: book_popularity.get(x[0], 0), reverse=True)

# Print the sorted books by popularity
print("{:<10} {:<50} {:<30} {:<15}".format("Book Number", "Title", "Author", "Days Borrowed"))
for book_number, book_info in sorted_books:
    days_borrowed = book_popularity.get(book_number, 0)
    print("{:<10} {:<50} {:<30} {:<15}".format(book_number, book_info['title'], book_info['author'], days_borrowed))

# Task 2: Test the genre popularity analysis
print("\nTask 2: Genre Popularity Analysis\n")
genre_popularity, genre_book_count = generate_genre_popularity_report(sample_books_data, sample_loans_data)
sorted_genre_report = sorted(genre_popularity.items(), key=lambda x: x[1], reverse=True)

# Print the sorted genre popularity report
print("{:<30} {:<15} {:<15}".format("Genre", "Total Days Borrowed", "Number of Books"))
for genre, popularity in sorted_genre_report:
    book_count = genre_book_count.get(genre, 0)
    print("{:<30} {:<15} {:<15}".format(genre, popularity, book_count))

# Task 3: Test loan statistics calculation
print("\nTask 3: Loan Statistics\n")
total_loan_days, total_loan_count, average_loan_days, \
total_late_loan_days, total_late_loan_count, average_late_loan_days, percentage_returned_late = \
    calculate_loan_statistics(sample_loans_data)

# Print loan statistics
print(f"The total loan days: {total_loan_days}")
print(f"The total loan count: {total_loan_count}")
print(f"The average loan days: {average_loan_days:.1f}")
print(f"The total late loan days: {total_late_loan_days}")
print(f"The total late loan count: {total_late_loan_count}")
print(f"The average late loan days: {average_late_loan_days:.1f}")
print(f"The percentage proportion of books returned late: {percentage_returned_late:.2f}%")

```

## Task 1 - Analysis of Popularity Measure

Task 1 uses the total number of days each book is borrowed in the year 2023 to measure popularity. While this approach can provide some insights into the usage of books in the library, it has some potential flaws:

###1. Lack of Context:

It only considers the number of days a book is borrowed without considering other factors such as the number of copies of a book available in the library. A book that has a single copy may have a lower total borrowed days because it was unavailable due to being borrowed by one person for an long period of time and this would skew the popularity results in favour of books that has several copies.

###2. Uniform Loan Period:
The measure assumes the loan period for all books is 14 days. However, in reality, different books may have different loan periods.

### 3. Ignoring Member Behavior:
It doesn't take into account member behavior. Some members might frequently borrow and return books, contributing to a book's high popularity, while others might prefer to keep books for longer periods.

### 4. Quality vs. Popularity:
Popularity does not necessarily correlate with the quality or significance of a book.

### 5. No Consideration of Late Returns:
The measure does not distinguish between books returned on time and those returned late. Late returns could be due to various reasons, and they might skew the popularity measure.

## Opinion on Popularity Measure:

The popularity measure used in Task 1 can serve as a basic indicator of which books were in demand during a particular year. However, for a more extensive understanding of a book's popularity, it would be useful to consider other factors such as the number of copies available, historical loan data, and member preferences.

Ultimately, while this measure is a useful starting point, it should not be the sole criterion for assessing the value or significance of books in a library. A more nuanced analysis that considers various factors would produce a more accurate picture.

```python
# Task 2: Define functions to analyze book genres

# Function to read books.csv into a dictionary with genre information
def read_books_csv_with_genres(file_path):
    """
    Read the books.csv file and store book information in a dictionary including genres.

    Args:
        file_path (str): The path to the books.csv file.

    Returns:
        dict: A dictionary containing book information with book numbers as keys, including genre.
    """
    books_data = {}
    with open(file_path, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            books_data[row['Number']] = {
                'title': row['Title'],
                'author': row['Author'],
                'genre': row['Genre']
            }
    return books_data

# Function to generate a popularity report of book genres
def generate_genre_popularity_report(books_data, valid_loans):
    """
    Generate a popularity report of book genres based on loan data.

    Args:
        books_data (dict): A dictionary containing book information.
        valid_loans (list): A list of dictionaries containing valid loan information.

    Returns:
        tuple: A tuple containing genre popularity and genre book count dictionaries.
    """
    genre_popularity = {}
    genre_book_count = {}

    # Calculate popularity and count for each genre
    for loan in valid_loans:
        book_number = loan['book_number']
        genre = books_data.get(book_number, {}).get('genre', 'Unknown')
        days_borrowed = loan['date_of_return'] - loan['date_of_loan']

        # Update genre popularity
        if genre not in genre_popularity:
            genre_popularity[genre] = days_borrowed
        else:
            genre_popularity[genre] += days_borrowed

        # Count books in each genre
        if genre not in genre_book_count:
            genre_book_count[genre] = 1
        else:
            genre_book_count[genre] += 1

    return genre_popularity, genre_book_count

# Main code for Task 2

# Read books data from CSV with genres
books_data = read_books_csv_with_genres(books_file_path)
# Generate genre popularity report based on loan data
genre_popularity, genre_book_count = generate_genre_popularity_report(books_data, valid_loans)

# Sort genre popularity report by popularity in reverse order
sorted_genre_report = sorted(genre_popularity.items(), key=lambda x: x[1], reverse=True)

#Collate results for Task 2
results['genre_popularity'] = genre_popularity
results['genre_book_count'] = genre_book_count

# Print the report

# Title for the report
task2_title = "Popularity Report of Book Genres Borrowed in 2023"

# Print the title box for Task 2
task2_title_box = create_title_box(task2_title)
print(task2_title_box)

# Create a list of lists containing the data to be tabulated
genre_table_data = []
genre_table_data.append(["Genre", "Total Days Borrowed", "Number of Books"])

for genre, popularity in sorted_genre_report:
    book_count = genre_book_count.get(genre, 0)
    genre_table_data.append([genre, popularity, book_count])

# Use tabulate to format the genre table and print it
print(tabulate(genre_table_data, headers="firstrow", tablefmt="fancy_grid"))
```

```python
# Task 3: Calculate loan statistics

# Function to calculate loan statistics
# Set the late threshold to 14 days because a book returned after 19 days is considered 5 days late
def calculate_loan_statistics(valid_loans, late_threshold=14):
    """
    Calculate loan statistics including total loan days, average loan days, and late loan statistics.

    Args:
        valid_loans (list): A list of dictionaries containing valid loan information.
        late_threshold (int): The number of days after which a book is considered late.

    Returns:
        tuple: A dictionary containing various loan statistics.
    """
    total_loan_days = 0
    total_loan_count = len(valid_loans)
    total_late_loan_days = 0
    total_late_loan_count = 0

    # Calculate total loan days and total late loan days
    for loan in valid_loans:
        days_borrowed = loan['date_of_return'] - loan['date_of_loan']
        total_loan_days += days_borrowed
        if days_borrowed > late_threshold:  # Books returned after 14 days are considered late
            total_late_loan_days += (days_borrowed - late_threshold)
            total_late_loan_count += 1

    # Calculate average loan days and average late loan days
    average_loan_days = total_loan_days / total_loan_count if total_loan_count > 0 else 0
    average_late_loan_days = total_late_loan_days / total_late_loan_count if total_late_loan_count > 0 else 0

    # Calculate the percentage proportion of books returned late
    percentage_returned_late = (total_late_loan_count / total_loan_count) * 100 if total_loan_count > 0 else 0

    return {
        'total_loan_days': total_loan_days,
        'total_loan_count': total_loan_count,
        'average_loan_days': average_loan_days,
        'total_late_loan_days': total_late_loan_days,
        'total_late_loan_count': total_late_loan_count,
        'average_late_loan_days': average_late_loan_days,
        'percentage_returned_late': percentage_returned_late
    }
# Main code for Task 3
loan_statistics = calculate_loan_statistics(valid_loans, late_threshold=14)

# Collate results for Task 3
results['loan_statistics'] = loan_statistics


# Print the loan statistics

# Title for Task 3
task3_title = "Loan Statistics for 2023"

# Print the title box for Task 3
task3_title_box = create_title_box(task3_title)
print(task3_title_box)

# Create a list of lists containing the data to be tabulated
loan_stats_table_data = [
    ["Statistic", "Value"],
    ["Total Loan Days", loan_statistics['total_loan_days']],
    ["Total Loan Count", loan_statistics['total_loan_count']],
    ["Average Loan Days", f"{loan_statistics['average_loan_days']:.1f}"],
    ["Total Late Loan Days", loan_statistics['total_late_loan_days']],
    ["Total Late Loan Count", loan_statistics['total_late_loan_count']],
    ["Average Late Loan Days", f"{loan_statistics['average_late_loan_days']:.1f}"],
    ["Percentage Returned Late", f"{loan_statistics['percentage_returned_late']:.2f}%"]
]

# Use tabulate to format the loan statistics table and print it
print(tabulate(loan_stats_table_data, headers="firstrow", tablefmt="fancy_grid"))

# Unmount Google Drive
drive.flush_and_unmount()
```
